```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline

import seaborn as sns
```

```python
#Read in Insurance cvs.
df = pd.read_csv("insurance.csv")
#Print head.
df.head(3)
```

```python
#Make a scatter plot from age and charges, define x as age and y as charges, reshape them.
x = df["age"].values.reshape(-1, 1)
y = df["charges"].values.reshape(-1, 1)

#Define linear regression by model name.
age_model = LinearRegression()
#Make data fit the model.
age_model.fit(x,y)
```

```python
#Now the data has been fitted - we can start making predictions, firstly i will make y predictions based on the x values I have.
pred_y = age_model.predict(x)

plt.figure()
#Plot an sns scatter plot with the reshaped data without the line of best fit.
sns.scatterplot(data=df, x="age", y="charges", hue="charges", legend=False)


#Label the plot.
plt.title("Age vs Insurance Charges")
#Change the x and y labels.
plt.ylabel("Charges")
plt.xlabel("Age")
```

```python
#Plot an sns scatter plot with the reshaped data with the line of best fit.
sns.scatterplot(data=df, x="age", y="charges", hue="charges", legend=False)

#Change the x and y labels.
plt.ylabel("Charges")
plt.xlabel("Age")
#Label the plot.
plt.title("Age vs Insurance Charges with Line of Best Fit")


#The line of best fit is in green.
plt.plot(x, pred_y, color="darkseagreen", linewidth=2)
```

There is not a very strong strong relationship between age and insurance charges. However, I will write a prediction for different.

```python
#Create a unique x value - I will do 80.
uni_x = [[80]]
#Create a prediction with the unique value.
x_pred = np.append(x, uni_x).reshape(-1,1)
#Create a y predition with the x predicion.
y_pred = age_model.predict(x_pred)

#Plot an sns scatter plot with the reshaped data with the line of best fit.
sns.scatterplot(data=df, x="age", y="charges", hue="charges", legend=False)

#Reformat the age predictions.
age_charg = age_model.predict(uni_x); age_charg = str(age_charg); age_charg = age_charg.replace("[[", "").replace("]]", " ")




#Change the x and y labels.
plt.ylabel("Charges")
plt.xlabel("Age")
#The line of best fit is in green, and made from the predictions.
plt.plot(x_pred, y_pred, color="darkseagreen", linewidth=2)
#Label the plot.
plt.title("Age vs Insurance Charges: Prediction for 80 Years")


#Print a prediction of what the insurance charge would be for someone ages 80.
print("A person aged 80 year's insurance should charge:",age_charg[:9])
```
